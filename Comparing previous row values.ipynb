In each example, we are comparing the value one row of the **col1** column with the value in the row immediately preceding it.

### Create dataframe

```python
import pandas as pd
data={'col1':[1,3,3,1,2,3,2,2]}
df=pd.DataFrame(data,columns=['col1'])
df
```

### Equal to ==

```python
df['=='] = df['col1'].diff().eq(0)
df
```

### Greater than >

```python
df.drop('==', axis=1, inplace=True)
```

```python
df['>'] = df['col1'].diff().gt(0)
df
```

### Less than <

```python
df.drop('>', axis=1, inplace=True)
```

```python
df['<'] = df['col1'].diff().lt(0)
df
```

### Greater than OR Equal to > =

```python
df.drop('<', axis=1, inplace=True)
```

```python
df['> ='] = df['col1'].diff().ge(0)
df
```

### Less than OR Equal to < =

```python
df.drop('> =', axis=1, inplace=True)
```

```python
df['< ='] = df['col1'].diff().le(0)
df
```

### Equal to: (0 or 1)

```python
df.drop('< =', axis=1, inplace=True)
```

```python
import numpy as np

df['=='] = np.where(df['col1'].diff().eq(0), 1, 0)
df
```

```python
df.drop('==', axis=1, inplace=True)
```

```python
df['diff'] = df['col1'].diff()
df
```
